```python
# STEP 1: Install openpyxl (needed to read Excel files)
!pip install openpyxl

# STEP 2: Import pandas
import pandas as pd

# STEP 3: Load your uploaded Excel file
df = pd.read_excel('/content/Train.xlsx')

# STEP 4: View the first few rows
df.head()
# Group by Date and count how many crimes happened per day
crime_per_day = df.groupby('Date').size().reset_index(name='Crime_Count')

# Show first few rows
crime_per_day.head()
# Rename columns to fit Prophet's requirements
crime_per_day.rename(columns={"Date": "ds", "Crime_Count": "y"}, inplace=True)

# Confirm the new column names
crime_per_day.head()
!pip install prophet
from prophet import Prophet

model = Prophet()
model.fit(crime_per_day)
# Create future dataframe (predict for 180 more days)
future = model.make_future_dataframe(periods=180)

# Predict future crime
forecast = model.predict(future)

# Show last few predicted values
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
# Plot the forecasted values
from prophet.plot import plot_plotly
plot_plotly(model, forecast)







```
